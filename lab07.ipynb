```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sb
%matplotlib inline

diamond=pd.read_csv("/Diamonds Prices2022.csv")
print(diamond.shape)
print(diamond.head)
```

```python
diamond.isnull().sum()
```

```python
diamond['carat'].value_counts()
```

```python
sb.countplot(data=diamond, x='depth');
```

```python
sb.color_palette()
```

```python
sb.countplot(data=diamond, x='price')
```

```python
depth_order=diamond['depth'].value_counts().index
depth_order
```

```python
sb.countplot(data=diamond, x='depth', order=depth_order);
```

```python

sorted_counts=diamond['depth'].value_counts()
plt.pie(sorted_counts, labels= sorted_counts.index, startangle=90, counterclock=False);
plt.axis('square');
```

```python

sorted_counts=diamond['table'].value_counts()
plt.pie(sorted_counts, labels= sorted_counts.index, startangle=90, counterclock=False);
plt.axis('square');
```
